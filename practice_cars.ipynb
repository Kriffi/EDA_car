# Практическая работа: Подготовка данных и первая модель классификации

Датасет: подержанные автомобили в США

```python
import pandas as pd
import numpy as np
```

## 1. Загрузка датасета и анализ пропусков

```python
df = pd.read_csv('vehicles_dataset.csv')

print(df.info())
print('\nПропуски по колонкам:')
print(df.isna().sum())

df.head()
```

## 2. Обработка пропусков в year

```python
df['year'] = df['year'].fillna(df['year'].median())
```

## 3. Обработка пропусков в fuel

```python
df['fuel'] = df['fuel'].fillna(df['fuel'].mode()[0])
```

## 4. Обработка пропусков в model

```python
df['model'] = df['model'].fillna('unknown')
```

## 5. Обработка пропусков в title_status

```python
df['title_status'] = df['title_status'].fillna(df['title_status'].mode()[0])
```

## 6. Преобразование типа year

```python
df['year'] = df['year'].astype(int)
```

## 7. Преобразование категориальных переменных

```python
categorical_cols = ['fuel', 'model', 'title_status', 'manufacturer', 'condition']

for col in categorical_cols:
    if col in df.columns:
        df[col] = df[col].astype('category')

df.dtypes
```

## 8. Поиск выбросов в year (IQR)

```python
Q1 = df['year'].quantile(0.25)
Q3 = df['year'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['year'] < lower_bound) | (df['year'] > upper_bound)]

print('Нижняя граница:', lower_bound)
print('Верхняя граница:', upper_bound)
print('Количество выбросов:', outliers.shape[0])
```

## (Необязательно) Сохранение обработанного датасета

```python
df.to_csv('vehicles_cleaned.csv', index=False)
```
